```python
data = data.strip()
data
```

```python
def test_removal(i, polymer):
    c1 = polymer[i]
    try:
        c2 = polymer[i+1]
    except IndexError:
        return False
    
    return c1.lower() == c2.lower() and c1.isupper() != c2.isupper()

def reduce_polymer(polymer):
    skip = False
    for i, c in enumerate(polymer):
        if skip:
            skip = False
            continue
        if test_removal(i, polymer):
            skip = True
        else:
            yield c
            

        
```

```python
polymer = 'aaABbA'

test_removal(3, polymer)
print(''.join(reduce_polymer(polymer)))
```

```python
def react_polymer(polymer):
    old_polymer = None
    while old_polymer != polymer:
        old_polymer = polymer
        polymer = ''.join(reduce_polymer(polymer))
    return polymer
```

```python
results = {u:0 for u in units}
```

```python
for u in units:
    removed_polymer = first_reacted_polymer.replace(u.upper(), '').replace(u.lower(), '')
    reduced_polymer = react_polymer(removed_polymer)
    results[u] = len(''.join(reduced_polymer))
```
