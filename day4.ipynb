```python
import re
import datetime
from collections import defaultdict
import numpy as np
```

```python
with open('day4.dat', 'r') as f:
    data = f.read()
```

```python
lines = data.split('\n')
lines[8]
```

```python
def get_date_index(line):
    return line.find('['), line.find(']')

def get_time(line):
    start, end = get_date_index(line)
    date = line[start:end].strip('[]')

    
    date, time = date.strip('[]').split()
    year, month, day = map(int, date.split('-'))
    hour, minute = map(int, time.split(':'))
    
    return datetime.datetime(year, month, day, hour, minute)
    

def get_string(line):
    _, end = get_date_index(line)
    return line[end+1:].strip()
    
def get_data(line):
    return get_time(line), get_string(line)

get_date_index(lines[8])
get_time(lines[8])
get_string(lines[8])
get_data(lines[8])
```

```python
data = [get_data(line) for line in lines if len(line) > 5]
```

```python
def get_guard_number(event):
    guard, nr, _, _ = event.split()
    nr = int(nr.strip('#'))
    return nr



def process_data(data):
    guard_sleeps = defaultdict(lambda: np.zeros(60))
    for time, event in data:
        if 'Guard' in event:
            nr = get_guard_number(event)
            current_guard = nr
        if 'falls asleep' in event:
            start = time.minute
        if 'wakes up' in event:
            end = time.minute
            guard_sleeps[current_guard][start:end] += 1
    return guard_sleeps
            
            

guard_sleep = process_data(data)
```

```python
def sum_sleep(guard_sleep):
    return {nr:sleeping.sum() for nr, sleeping in guard_sleep.items()}

def get_sleepiest_guard(guard_sleep):
    summed_sleep = sum_sleep(guard_sleep)
    sleepiest = max(summed_sleep, key=summed_sleep.get)
    return sleepiest

def get_sleepiest_minute(guard_sleep, guard):
    sleep_times = dict(enumerate(guard_sleep[guard]))
    return max(sleep_times.values()), max(sleep_times, key=sleep_times.get)

def get_answer(guard_sleep):
    guard = get_sleepiest_guard(guard_sleep)
    total, minute = get_sleepiest_minute(guard_sleep, guard)
    return guard * minute


def guards_sleepist_minutes(guard_sleep):
    return {guard: get_sleepiest_minute(guard_sleep, guard) for guard in guard_sleep.keys()}

def strat2_sleepiest_guard(guard_sleep):
    sleepiest_minutes = guards_sleepist_minutes(guard_sleep)
    return max(sleepiest_minutes, key=sleepiest_minutes.get)

def get_answer2(guard_sleep):
    sleepiest_minutes = guards_sleepist_minutes(guard_sleep)
    guard = strat2_sleepiest_guard(guard_sleep)
    return guard * sleepiest_minutes[guard][1]

get_answer(guard_sleep)
strat2_sleepiest_guard(guard_sleep)
guards_sleepist_minutes(guard_sleep)
get_answer2(guard_sleep)
```
