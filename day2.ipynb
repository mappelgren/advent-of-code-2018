```python
def x_counts(line, x):
    cnt = Counter()
    for c in line.strip():
        cnt[c] += 1
    total = 0
    for c in cnt.elements():
        if cnt[c] == x:
            total += 1
    return total

def three_counts(line):
    return x_counts(line, 3)
    
def two_counts(line):
    return x_counts(line, 2)

x_counts('oeambtcgjqnzhgkdylfapoiusr\n', 2)
```

```python
def diff(id1, id2):
    total = 0
    for c1, c2 in zip(id1, id2):
        if c1.strip() != c2.strip():
            total += 1
    return total

def find_answer(data):
    for id1 in data:
        for id2 in data:
            if diff(id1, id2) == 1:
                return id1.strip(), id2.strip()
id1, id2 = find_answer(data)
```
